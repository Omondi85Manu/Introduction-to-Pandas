### merging ,Joining and concatenating

```python
df3 =pd.DataFrame({'A': ['A0','A1','A2','A3'],
                  'B': ['B0','B1','B2','B3'],
                  'C': ['C0','C1','C2','C3'],
                  'D': ['DO','D1','D2','D3']},
                  index=[0,1,2,3] )
df3
```

```python
df4 =pd.DataFrame({'A': ['A4','A5','A6','A7'],
                  'B': ['B4','B5','B6','B7'],
                  'C': ['C4','C5','C6','C7'],
                  'D': ['D4','D5','D6','D7']},
                  index=[4,5,6,7] )
df4
```

# Concatenation
You can use **pd.concat** and pass in the list of DataFrames to concatenate together


```python
pd.concat([df3,df4])
```

```python
# Concatenate the column
pd.concat([df3,df4], axis=1)
```

### Merging
The **merge** function allows you to merge Dataframes together using as similar logic as merging in SQL Table together

```python
left =pd.DataFrame({'key': ['A0','A1','A2','A3'],
                  'B': ['B0','B1','B2','B3'],
                  'C': ['C0','C1','C2','C3'],
                  'D': ['DO','D1','D2','D3']})
left
```

```python
df4 =pd.DataFrame({'key':['A0','A1','A2','A3'],
                  'B': ['B4','B5','B6','B7'],
                  'C': ['C4','C5','C6','C7'],
                  'D': ['D4','D5','D6','D7']}
                )
```

```python
pd.merge(left,right, how= 'inner',on='key')
```

```python
pd.merge(right,left, how= 'inner',on='key')
```

```python
left1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']}) 
```

```python
pd.merge(right1,left1, how= 'inner',on='key')
```

```python
#Load html file
import requests
from bs4 import BeautifulSoup
import pandas as pd

```

```python
# get the html page
page = requests.get("https://www.must.ac.ke/kuccps-2019-admission-list-2/")
soup = BeautifulSoup(page.text, 'lxml')
soup
```

```python
# get the table

table = soup.find('table')

headers = []

for rows in table.find_all('tr')[:1]:
    data = rows.find_all('td')
    titles = [td.text.strip() for td in data]
    headers.append(titles)
df =pd.DataFrame(columns = headers)
df
```

```python
# getting the table content

for row in table.find_all('tr')[1:]:
    row_data = row.find_all('td')
    content = [td.text for td in row_data]
    length = len(df)
    df.loc[length]=content
    
df
```
